# Which team one nationality won the most games?

```python
# sort by nationality first, then what team they were one, then if they won or not, then add it all together
#columns of intrest: winner, team_one_nationalities, team_two_nationalities
small_df = df[['winner', 'team_one_nationalities', 'team_two_nationalities']]
small_df.head(50)

d = {}
for index, row in small_df.iterrows():
    winner = row[0]
    t1 = row[1]
    t2 = row[2]
    #Fill the dictionary
    if winner == 1:
        if t1 in d:
            d[t1] = d[t1] + 1
        else:
            d[t1] = 1
    else:
        if t2 in d:
            d[t2] = d[t2] + 1
        else:
            d[t2] = 1
sorted_wins = sorted(d.items(), key=lambda x: x[1], reverse = True)
answer = sorted_wins[0]
type(answer)
print('The country with the most wins is', answer[0], 'with', answer[1], '')
```

The country with the most wins is INDIA with 484 wins!

# Which player has the most points overall?

```python
#useful columns: team_one_players, team_two_players, team_one_total_points, team_two_total_points
small_df = df[['team_one_players', 'team_two_players', 'team_one_total_points', 'team_two_total_points']]
small_df.head(5)
d = {}
for index, row in small_df.iterrows():
    t1_player = row[0]
    t2_player = row[1]
    t1_points = row[2]
    t2_points = row[3]
    #Handle t1
    if t1_player in d:
        d[t1_player] = d[t1_player] + t1_points
    else:
        d[t1_player] = t1_points
    
    #Handle t2
    if t2_player in d:
        d[t2_player] = d[t2_player] + t2_points
    else:
        d[t2_player] = t2_points
        
        
sorted_points = sorted(d.items(), key=lambda x: x[1], reverse = True)
answer = sorted_points[0]
print('The player with the most points overall is', answer[0], 'with', answer[1], 'total points!')
```

The player with the most points overall is Tien Chen Chou with 6259 total points!
